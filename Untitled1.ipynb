```python
import csv
product_details ={}
with open("/product details.csv","r")as file:
  reader=csv.DictReader(file)
  print(file)
  for row in reader:
    product_id = row["PRODUCT ID"]
    details = [row['PRODUCT NAME'],int(row["PRODUCT PRICE"]),int(row["PRODUCT QUANTITY"])]
    product_details[product_id] = details
  for product_id,details in product_details.items():
    print(f"PRODUCT ID:{product_id}") 
    print(f"PRODUCT NAME:{details[0]}")
    print(f"PRODUCT PRICE:{details[1]}")
    print(f"PRODUCT QUANTITY:{details[2]}")
    print()
```
